#### Scenario 04: Voting System (Annual Employee Recognition Awards)


```python
print("-------- Annual Employee Recognition Awards --------")
award_categories = ["Employee of the Year", "Team Player of the Year", "Innovation Award"]
print(award_categories)
candidates = ["Candidate A", "Candidate B", "Candidate C"]
print(candidates)
```

```python
voters = int(input("Enter the Number of voters participating: "))
```

```python
voting_result = {}
for award in award_categories:
    print("-------------------------------------")
    print(f"\nAward Category : {award}")
    for candidate in range(len(candidates)):
        print(f"{candidate+1}. {candidates[candidate]}")
    print("-------------------------------------")
    print("\nNOTE: 1 attempts Only")
    candidate_count = {}
    candidate1_count = 0
    candidate2_count = 0
    candidate3_count = 0
    for voter in range(1,voters+1):
        vote = int(input(f"\nVoter {voter}: Cast your Vote for {award} : "))
        if vote > 0 and vote <=3:
            print("\nSuccesfully Recorded Your Vote")
            print("-------------------------------------")
            if vote == 1:
                candidate1_count +=1
            elif vote == 2:
                candidate2_count +=1
            elif vote == 3:
                candidate3_count +=1
            candidate_count[candidates[0]] = candidate1_count
            candidate_count[candidates[1]] = candidate2_count
            candidate_count[candidates[2]] = candidate3_count
        else:
            print("\nInvalid Vote! Try next time")
    voting_result[award] = candidate_count
print(voting_result)
```

##### Results Display:

Display the voting results, showing the number of votes each candidate received in each category.

```python
for award, value in voting_result.items():
    print(f"\n{award}\n")
    for candidate, vote in value.items():
        print(f"{candidate} got {vote} votes")
    winner = candidate[max(value.values())]
    print("\n---------------------------------")
    print(f"Winner of the {award} is: {winner}")
    print("---------------------------------\n")
```
